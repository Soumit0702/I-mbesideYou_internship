```python
csv_file_1 = pd.read_csv('/content/Emotion_score.csv')
csv_file_2 = pd.read_csv('/content/gaze_score_updated.csv')
csv_file_3 = pd.read_csv('/content/transcript_scores.csv')

merged_df = csv_file_1.merge(csv_file_2, on=csv_file_1.columns[0], how='inner') \
                      .merge(csv_file_3, on=csv_file_1.columns[0], how='inner')
merged_df.to_csv('Score_total.csv', index=False)

print("CSV files merged successfully based on the first primary column!")
```

```python
import pandas as pd

file_path = '/content/Score_total.csv'
score_data = pd.read_csv(file_path)

weights = {
    'Emotion_Score': 0.5,
    'Final Score_gaze': 0.3,
    'transcript_Score': 0.2
}

score_data['Weighted_Score'] = (score_data['Emotion_Score'] * weights['Emotion_Score'] +
                                score_data['Final Score_gaze'] * weights['Final Score_gaze']+
                                score_data['transcript_Score'] * weights['transcript_Score'])

threshold = 15

selected_candidates = score_data[score_data['Weighted_Score'] >= threshold]

print(selected_candidates[['Candidate', 'Weighted_Score']])
```

```python
data = selected_candidates.to_csv("Selected_Candidates_list.csv")
```
